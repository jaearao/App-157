```python
import numpy as np
import matplotlib.pyplot as plt
import random

# set up x-axis
time = np.linspace(0, 100, 1000)
# number of iterations:
iter = 1000
# initial population:
pop_init = 100
# rate constant:
k = 1/6
f_pop = np.arange(pop_init, -1, -1)

# storing average:
average = np.zeros(pop_init+1)




def Rate(n):
    rate = k * n
    return(rate)

for i in range(iter):
    pop = pop_init
    mu, sigma = 0.5, 0.2
    dt = np.array([0])
    while(pop != 0):
        dt = np.append(dt, random.gauss(mu, sigma))
        pop -= 1
    pop = pop_init
    for t in range(1, len(dt)):
        dt[t] = dt[t-1] + 2*dt[t]/Rate(pop)
        pop -= 1
    
    plt.plot(dt, f_pop, 'gray', linewidth = 1, alpha = 0.2)
    for j in range(len(dt)):
        average[j] += dt[j]

average = average / iter

plt.plot(average, f_pop, label = "Average")
plt.title("Continuous Radioactive Decay")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Population")
```
